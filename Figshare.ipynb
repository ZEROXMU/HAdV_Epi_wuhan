# Fig 2C

```python
d3 = pd.read_excel('Figshare.xlsx', sheet_name='Fig2C')
d3.head(2)
```

```python
plt.rc('font', family='Times New Roman')
plt.rcParams.update({'font.size':8})
plt.rcParams.update({'font.weight': 'bold'})

colors = ['#fc4410','#186CB6','#784D9C','#F1AC71','#B92885']

titles = [
    'Positive rate',
   'Infected',
    'Hospitalized',
    'Severe',
    'Death',]


fig, axs = plt.subplots(1, 5, figsize=(14, 3))
axs = axs.flatten()

# Loop through J1['TYPE'].unique()
for i, (x_value, title, color) in enumerate(zip(d3['TYPE'].unique(), titles, colors)): 
    subset = d3[d3['TYPE'] == x_value]
    
    ax1 = axs[i]
    
    # Plot bar chart
    ax1.bar(subset['TIME'], subset['value'], color=color, alpha=0.4, label='Original data', width=24)
    
    # Plot line chart
    ax1.plot(subset['TIME'], subset['model_fit'], color=color, label='Model fit', linewidth=1.2)
    
    # Mark and text
    for idx, row in subset.iterrows():
        if row['MPC'] == 'JPR':
            ax1.plot(row['TIME'], row['model_fit'], 'ro', markersize=5)
        
    non_jpr_apcs = [apc for apc in subset['MPC'].unique() if apc != 'JPR']

    ax1.set_title(f'{title}', fontdict={'fontsize': 14, 'fontweight': 'bold'})
    ax1.tick_params(axis='x', rotation=90)
    
    # Set y-limits based on subplot index
    if i == 0:
        ax1.set_ylim(0, 0.07)
    elif i == 1:
        ax1.set_ylim(0, 3200)
    elif i == 2:
        ax1.set_ylim(0, 1600)
    elif i == 3:
        ax1.set_ylim(0, 40)
    elif i == 4:
        ax1.set_ylim(0, 5)
    
    specific_timestamp = pd.Timestamp('2023-05-01 00:00:00')
    y_min, y_max = ax1.get_ylim()
    y_interval = (y_max - y_min) / (len(non_jpr_apcs) + 15)
    
    # Adjust text position slightly to the left
    
    adjusted_timestamp = specific_timestamp + pd.Timedelta(days=40)
  
        
    for j, apc_value in enumerate(non_jpr_apcs):
        ax1.text(adjusted_timestamp, y_max - (j) * y_interval, 
                 f'MPC {j + 1} = {apc_value}', fontsize=8, ha='right', va='top')

# Adjust layout to make space for the legend
plt.tight_layout(rect=[0, 0.1, 1, 0.95])

#plt.savefig(f'C:\\Users\\xmu0412\\Desktop\\RES_xbd\\JPR.pdf', dpi=450,bbox_inches='tight') 
plt.show()
```

```python
df = pd.read_excel('Figshare.xlsx', sheet_name='Fig3B')
df.head(2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25, 5))
plt.subplots_adjust(wspace=0.3)

# 2. 动态人口基数函数
def get_population(date_str):
    """根据日期返回对应人口基数"""
    return 13809100 if pd.to_datetime(date_str).year == 2024 else 13774000

# 3. 循环绘制每个参数
for i, (ax, para_name) in enumerate(zip(axes, df['Para'].unique())):
    # 3.1 筛选当前参数的数据
    para_df = df[df['Para'] == para_name].copy()
    
    # 3.2 计算发病率（动态人口基数）
    para_df['Incidence'] = para_df.apply(
        lambda row: row['Sensitivity_Mean'] * 100000 / get_population(row['Date']),
        axis=1
    )
    para_df['Lower'] = para_df.apply(
        lambda row: (row['Sensitivity_Mean'] - row['Sensitivity_Std']) * 100000 / get_population(row['Date']),
        axis=1
    )
    para_df['Upper'] = para_df.apply(
        lambda row: (row['Sensitivity_Mean'] + row['Sensitivity_Std']) * 100000 / get_population(row['Date']),
        axis=1
    )
    
    # 3.3 绘制主曲线
    ax.plot(para_df['Date'], para_df['Incidence'], 
            label='Mean', color='k', lw=2)
    
    # 3.4 绘制置信区间（修正后的关键部分）
    ax.fill_between(para_df['Date'].values,  # 转换为numpy数组
                   para_df['Lower'].values,
                   para_df['Upper'].values,
                   alpha=0.3, color='#B92885', label='SD')
    
    # 3.5 设置子图属性
    ax.set_title(para_name, fontweight='bold')
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax.tick_params(axis='x', rotation=90)
    ax.set_ylim(0, 140)
    
    # 只在第一个子图显示图例
    if i == 0:
        ax.legend(loc='upper right')

# 4. 共享轴标签
fig.text(0.5, 0.02, 'Date', ha='center', fontweight='bold')
plt.tight_layout()
#plt.savefig(f"C:\\Users\\xmu0412\\Desktop\\RES_xbd\\敏感性分析.pdf", dpi=450, bbox_inches='tight') 
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats


plt.rc('font', family='Times New Roman')
plt.rcParams.update({'font.size': 14, 'font.weight': 'bold'})

shapiro_result = stats.shapiro(df['R0'])
ks_result = stats.kstest(df['R0'] , 'norm', args=(df['R0'].mean(), df['R0'].std()))

# 绘制直方图
plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.hist(df['R0'] , bins=30, color='#186CB6', edgecolor='black')
plt.xlabel('R0', fontdict={'fontsize': 14, 'fontweight': 'bold'})
plt.ylabel('Frequency', fontdict={'fontsize': 14, 'fontweight': 'bold'})

# 绘制 Q-Q 图
plt.subplot(1, 2, 2)
stats.probplot(df['R0'] , dist="norm", plot=plt)

# 添加测试结果到 Q-Q 图
sw_pvalue = f"{shapiro_result.pvalue:.3f}" + ("*" if shapiro_result.pvalue < 0.05 else "")
ks_pvalue = f"{ks_result.pvalue:.3f}" + ("*" if ks_result.pvalue < 0.05 else "")

plt.text(0.05, 0.95, f"S-W Stat: {shapiro_result.statistic:.3f}, p: {sw_pvalue}",
         transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color='k', fontweight='bold')
plt.text(0.05, 0.90, f"K-S Stat: {ks_result.statistic:.3f}, p: {ks_pvalue}",
         transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color='k', fontweight='bold')
plt.xlabel('Theoretical quantiles', fontdict={'fontsize': 14, 'fontweight': 'bold'})
plt.ylabel('Ordered Values', fontdict={'fontsize': 14, 'fontweight': 'bold'})
plt.title("") 
plt.tight_layout()
#plt.savefig(f"C:\\Users\\xmu0412\\Desktop\\RES_xbd\\R0分布.pdf", dpi=450, bbox_inches='tight') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

plt.rc('font', family='Times New Roman')
plt.rcParams.update({'font.size': 15})
plt.rcParams['font.weight'] = 'bold'



columns_to_check = ['T', 'Tx', 'Tn', 'RH', 'P', 'WS', 'PRCP', 'AQI', 'PM2.5',
               'PM10', 'SO2', 'NO2', 'CO', 'O3', 'CBMI','BSI']

# Define colors
colors = ['#186CB6'] * 7 + ['#784D9C'] * 7 + ['#F1AC71'] * 2

# 创建子图
fig, axes = plt.subplots(nrows=16, ncols=2, figsize=(12, 48))
axes = axes.flatten()

for i, column in enumerate(columns_to_check):
     
    sns.histplot(sliced_data[column], kde=True, ax=axes[2*i], color=colors[i])
    axes[2*i].set_title(f'Histogram of {columns_to_check [i]}', fontdict={'fontsize': 15, 'fontweight': 'bold'})
    axes[2*i].set_ylabel('Frequency', fontdict={'fontsize': 14, 'fontweight': 'bold'})
    axes[2*i].set_xlabel('Value', fontdict={'fontsize': 14, 'fontweight': 'bold'})

    # 绘制QQ图
    stats.probplot(sliced_data[column], dist="norm", plot=axes[2*i + 1])
    axes[2*i + 1].get_lines()[0].set_color(colors[i])  # Set color for QQ plot
    axes[2*i + 1].set_title(f'QQ Plot of {columns_to_check [i]}', fontdict={'fontsize': 15, 'fontweight': 'bold'})
    axes[2*i + 1].set_ylabel('Ordered Values', fontdict={'fontsize': 14, 'fontweight': 'bold'})
    axes[2*i + 1].set_xlabel('Theoretical Quantiles', fontdict={'fontsize': 14, 'fontweight': 'bold'})
    # --- 正态性检验 ---
    ks_stat, ks_p = stats.kstest(sliced_data[column], 'norm', 
                                args=(sliced_data[column].mean(), sliced_data[column].std()))
    sw_stat, sw_p = stats.shapiro(sliced_data[column])
    
    # 标记显著性（p<0.05时加星号）
    ks_sig = '*' if ks_p < 0.05 else ''
    sw_sig = '*' if sw_p < 0.05 else ''
    
   
    axes[2*i + 1].text(0.05, 0.95, f'K-S stat: {ks_stat:.2f}, p: {ks_p:.3f}{ks_sig}\nS-W stat: {sw_stat:.2f}, p: {sw_p:.3f}{sw_sig}',
                       transform=axes[2*i + 1].transAxes, fontsize=14,
                       verticalalignment='top')

plt.tight_layout()
#plt.savefig(f"C:\\Users\\xmu0412\\Desktop\\RES_xbd\\无滞后变量正态性检验.pdf", dpi=450, bbox_inches='tight') 
plt.show()
```

# Fig S9

```python
from scipy.stats import spearmanr

start_date = '2022-12-01'
end_date = '2024-06-21'
columns_to_check = ['T', 'Tx', 'Tn', 'RH', 'P', 'WS', 'PRCP', 'AQI', 'PM2.5',
               'PM10', 'SO2', 'NO2', 'CO', 'O3', 'CBMI', 'BSI']


sliced_data = df.loc[start_date:end_date]

sliced_data.index = pd.to_datetime(sliced_data.index)

start_date1 = '2023-01-01'
end_date1 = '2024-05-21'

ince = (df['Number of cases']/ df['Population']).loc[start_date1:end_date1] * 100000
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr
columns_to_check = ['T', 'Tx', 'Tn', 'RH', 'P', 'WS', 'PRCP', 'AQI', 'PM2.5',
               'PM10', 'SO2', 'NO2', 'CO', 'O3', 'CBMI', 'BSI']
plt.rc('font', family='Times New Roman')
plt.rcParams.update({'font.size': 15})
plt.rcParams['font.weight'] = 'bold'


# Colors for the plots
colors = ['#186CB6'] * 7 + ['#784D9C'] * 7 + ['#F1AC71'] * 2

# Create a 5x4 subplot
fig, axes = plt.subplots(4, 4, figsize=(15, 15))
axes = axes.flatten()

# Initialize result DataFrame
res_df = pd.DataFrame(columns=['Column', 'Max_Corr', 'Lag'])

# Calculate correlation coefficients
for idx, col in enumerate(columns_to_check):
    if col in sliced_data.columns:  # Ensure the column exists
        max_corr = 0
        optimal_lag = 0
        lags = range(-30, 1)
        correlations = []
        p_values = []

        for i in lags:
            # Calculate the lagged correlation
            lagged_data = sliced_data[col].shift(i)
            aligned_data, aligned_lagged = lagged_data.align(ince, join='inner')
            corr, p_value = spearmanr(aligned_data, aligned_lagged)
            correlations.append(corr)
            p_values.append(p_value)

            # Update the maximum correlation and corresponding lag
            if abs(corr) > abs(max_corr):
                max_corr = corr
                optimal_lag = i

        # Save the result
        new_row = pd.DataFrame({'Column': [col], 'Max_Corr': [max_corr], 'Lag': [optimal_lag]})
        res_df = pd.concat([res_df, new_row], ignore_index=True)

        # Plotting on the subplot
        ax = axes[idx]
        ax.bar(lags, correlations, color=colors[idx], alpha=0.6)
        ax.set_ylim(-1, 1)
        ax.set_xlim(-31, 1)

        # Add stars for significant correlations
        for j, (corr, p) in enumerate(zip(correlations, p_values)):
            if p < 0.05:
                y_pos = correlations[j] if correlations[j] > 0 else correlations[j] - 0.03
                ax.text(lags[j], y_pos, '*', color='k', fontsize=8, ha='center', va='bottom' if correlations[j] > 0 else 'top')

        # Highlight the maximum correlation
        ax.scatter(optimal_lag, 0, color='#B92885', marker='v', s=60, zorder=20)
        ax.axhline(y=0, color='black', linestyle='--', label='r = 0')

        ax.set_title(f'{columns_to_check[idx]}', fontdict={'fontsize': 15, 'fontweight': 'bold'})
        ax.set_xlabel('Lags (days)', fontdict={'fontsize': 15, 'fontweight': 'bold'})
        ax.set_ylabel('Correlation coefficient', fontdict={'fontsize': 15, 'fontweight': 'bold'})

# Turn off the last subplot
#axes[-1].axis('off')

# Adjust layout
plt.tight_layout()

#plt.savefig(f"C:\\Users\\xmu0412\\Desktop\\RES_xbd\\变量纯滞后.pdf", dpi=450, bbox_inches='tight')
plt.show()
```
